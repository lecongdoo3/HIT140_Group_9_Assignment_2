## Feature engineering

### New variables

##### jitter-shimmer-difference ratio = Jitter(d.d.p) / Shimmer(d.d.a)

Both Jitter and Shimmer are pivotal for voice quality. So, we choose "jitter_shimmer_difference_ratio" because it combines variations between consecutive voice cycles, providing a comprehensive assessment of voice stability and instability, particularly vital in distinguishing Parkinson's symptoms.

AZADI, H., AKBARZADEH-T, M.-R., SHOEIBI, A. & KOBRAVI, H. R. 2021. Evaluating the Effect of Parkinson's Disease on Jitter and Shimmer Speech Features. Advanced biomedical research [Online], 10. Available: http://europepmc.org/abstract/MED/35127581
https://doi.org/10.4103/abr.abr_254_21
https://europepmc.org/articles/PMC8781904
https://europepmc.org/articles/PMC8781904?pdf=render [Accessed 2021].


```python
jitter_shimmer_difference_ratio = df.loc[:, 'jitter(ddp)'] / df.loc[:, 'shimmer(dda)']
jitter_shimmer_difference_ratio_df = pd.DataFrame(jitter_shimmer_difference_ratio, columns=['jitter_shimmer_difference_ratio'])
jitter_shimmer_difference_ratio_df
```

#####  Harmonicity Difference = hnr− nhr

The difference between the Harmonic-to-Noise ratio (HNR) and Noise-to-Harmonic ratio (NHR) can give insights into the balance of noise and harmonics in the voice. A larger difference might indicate better voice quality, while a smaller difference may indicate disturbances, potentially correlating with disease severity.

BENBA, A., JILBAB, A. & HAMMOUCH, A. 2016. Voice analysis for detecting patients with Parkinson's disease using the hybridization of the best acoustic features. International Journal on Electrical Engineering and Informatics, 8, 108.

FERRAND, C. T. 2002. Harmonics-to-noise ratio: an index of vocal aging. Journal of voice, 16, 480-487.

```python
harmonicity_difference = df.loc[:,'hnr'] - df.loc[:,'nhr']
harmonicity_difference_df = pd.DataFrame(harmonicity_difference, columns=['harmonicity_difference'])
harmonicity_difference_df
```

### Add new variables to make a new dataset

```python
# Combine the new variables with the old data to create a single dataset, getting it ready for analysis.
new_df = pd.concat((df, jitter_shimmer_difference_ratio_df, harmonicity_difference_df,), axis=1, ignore_index=False)
new_df
```

```python
# Save the new data to a new .csv file
new_df.to_csv('new_po2_data.csv', index=False, header=True)
```
